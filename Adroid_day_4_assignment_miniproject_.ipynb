```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop(['ShapeFactor3'], axis=1)
df.head(5)
```

```python
df=df.rename(columns={'ShapeFactor1':'SF1','ShapeFactor2':'SF2','ShapeFactor3':'SF3'})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['Area'],color='red')
```

```python
sns.boxplot(x=df['Perimeter'],color='green')
```

```python
class_convex_area = df.groupby('Class')['ConvexArea'].mean()

plt.figure(figsize=(5, 8))
plt.bar(class_convex_area.index, class_convex_area.values, color='yellow')
plt.title('Bar plot of ConvexArea for Each Class')
plt.xlabel('Class')
plt.ylabel('Mean ConvexArea')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=df, color='green')
plt.title('countplot of Class')
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['Class'],color='teal')
plt.title('Histogram of Class')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['MinorAxisLength'], y=df['Class'],color='brown')
plt.title('Scatter plot of MinorAxisLength vs Class')
```

```python
plt.figure(figsize=(5,8))
sns.lineplot(y=df['MajorAxisLength'], x=df['Class'],color='yellow')
plt.title('Line plot of MajorAxisLength vs Class')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Class'] = label_encoder.fit_transform(df['Class'])
print(df.head())
```
